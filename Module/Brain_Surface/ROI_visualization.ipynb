# Params

```python
hemisphere = "L"
atlas = "Brodmann"

resolution = 32

# ROI
roi_label_path = f"/mnt/sda2/Common_dir/Atlas/Surface/fs_LR_{resolution}/{atlas}/{hemisphere}_rois.npy"
roi_value_path = f"/mnt/sda2/Common_dir/Atlas/Surface/fs_LR_{resolution}/{atlas}/{hemisphere}_roi_values.npy"
roi_info_path = f"/mnt/sda2/Common_dir/Atlas/Surface/fs_LR_{resolution}/{atlas}/{hemisphere}_roi_vertex_info.json"

# Sulcus
sulcus_path = f"/mnt/sda2/Common_dir/Atlas/Surface/fs_LR_{resolution}/borders/{hemisphere}_sulcus.json"

# Surf
template_surface_path = f"/mnt/sda2/Common_dir/Atlas/Surface/fs_LR_{resolution}/fs_LR.{resolution}k.{hemisphere}.flat.surf.gii"
```

# Load Datas

```python
roi_labels = np.load(roi_label_path)
roi_values = np.load(roi_value_path)
with open(roi_info_path, 'rb') as file:
    roi_info = json.load(file)
```

```python
with open(sulcus_path, "r") as file:
    sulcus_data_info = json.load(file)
```

```python
temploate_surface_data = nb.load(template_surface_path)
vertex_data = temploate_surface_data.darrays[0].data
```

# Visualization - ROIs

```python
draw_surf_roi(roi_value_array = roi_values,
              roi_info = roi_info,
              surf_hemisphere = hemisphere)
```

```python
draw_surf_selectedROI(surf_roi_labels = roi_labels, 
                      roi_name = f"|Rt_insular|", 
                      surf_hemisphere = hemisphere, 
                      alpha = 0.3)
```

# Visualization - sulcus

```python
n_vertex, _ = vertex_data.shape

surf_ax = surf.plot.plotmap(data = np.zeros(n_vertex), 
                            surf = f"fs{resolution}k_{hemisphere}",
                            threshold = 0.01)

for sulcus_name in sulcus_data_info:
    surf_ax.plot(np.array(sulcus_data_info[sulcus_name])[:, 0], 
                 np.array(sulcus_data_info[sulcus_name])[:, 1], 
                 color = "white",  
                 linestyle='dashed')
```
