# Common Library

```python
import nibabel as nb
import matplotlib.pyplot as plt
import nibabel as nb
import numpy as np
```

# Functions

```python
def reference2imageCoord(xyz, affine):
    """
    change reference coordinate(MNI, RAS+) to image coordinate
    reference coordinate can be scanner coordinate or MNI coordinate...
    
    :param xyz: anatomical coordinate(np.array)
    :param affine: affine matrix(np.array)
    
    return image coordinate(np.array)
    """
    
    result = np.matmul(np.linalg.inv(affine), [xyz[0], xyz[1], xyz[2], 1])[0:3]
    result = np.ceil(result).astype(int) # note: This is ad-hoc process - (np.ceil)
    return result
```

```python
def show_nifti_Image(image_path,
                     mni_coords, 
                     figsize = (12, 5.5)):
    """
    Show nifti image
    
    :param image_path(string): image path 
    :param mni_coords(list): mni coordinate [x, y, z] ex) [55, -53, -18}]
    """
    # Load data
    img = nb.load(template)
    data = img.get_fdata()
    
    # Coordinate
    x_o, y_o, z_o = mni_coords[0], mni_coords[1], mni_coords[2]
    x, y, z = reference2imageCoord(mni_coords, affine = img.affine)
    
    # Slice - sagittal, coronal, axial
    sagittal = data[x,:,:].T
    sagittal_remove_bg = np.where(sagittal != 0, sagittal, np.max(sagittal))

    coronal = data[:,y,:].T
    coronal_remove_bg = np.where(coronal != 0, coronal, np.max(coronal))

    axial = data[:,:,z].T
    axial_remove_bg = np.where(axial != 0, axial, np.max(axial))

    # Figure
    f, axes = plt.subplots(1, 3, figsize = figsize)
    
    # Img
    axes[0].imshow(sagittal_remove_bg, origin='lower',cmap='gray')
    axes[1].imshow(coronal_remove_bg,origin='lower',cmap='gray')
    axes[2].imshow(axial_remove_bg,origin='lower',cmap='gray')
    
    # Title
    axes[0].set_title(f"x={x_o}")
    axes[1].set_title(f"y={y_o}")
    axes[2].set_title(f"z={z_o}")
    
    # Axis
    axes[0].invert_xaxis()
    
    # Axis - visible
    for i in range(3):
        axes[i].xaxis.set_visible(False)
        axes[i].yaxis.set_visible(False)
        axes[i].axis("off")
        
    return f, axes
```

# Main

```python
mni_coords = 55, -53, -18 # MNI coordinates(RAS+)
```

```python
template = "/usr/local/afni/abin/MNI152_2009_template.nii.gz"
```

```python
show_nifti_Image(image_path = template, mni_coords = mni_coords)
```
